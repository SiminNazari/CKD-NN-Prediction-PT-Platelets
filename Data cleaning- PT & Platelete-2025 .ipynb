```python
# Select and keep the desired columns in DataFrame and Removing others

import pandas as pd

# Assuming your DataFrame is named df
selected_columns = ['age', 'gendera', 'BMI', 'hypertensive', 'glucose', 'RBC', 'PH', 'EF', 'COPD',
                     'PT','Platelets','atrialfibrillation','CHD with no MI', 'Hyperlipemia', 'deficiencyanemias', 'depression']

df = df[selected_columns]
```

```python
# Load your dataset
#df_cleaned = pd.read_csv('path_to_your_file/df_cleaned.csv')  # Make sure to replace 'path_to_your_file' with the actual path to your CSV file

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Plot distributions of PT and Platelets
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot for PT
sns.histplot(df_cleaned['PT'].dropna(), ax=axes[0], bins=30, kde=True, color='skyblue')
axes[0].set_title('Distribution of PT (Prothrombin Time)')
axes[0].set_xlabel('PT (seconds)')
axes[0].set_ylabel('Frequency')

# Plot for Platelets
sns.histplot(df_cleaned['Platelets'].dropna(), ax=axes[1], bins=30, kde=True, color='salmon')
axes[1].set_title('Distribution of Platelets')
axes[1].set_xlabel('Platelets (x10^9/L)')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Plot distributions of age and BMI
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot for PT
sns.histplot(df_cleaned['age'].dropna(), ax=axes[0], bins=30, kde=True, color='red')
axes[0].set_title('Distribution of age')
axes[0].set_xlabel('age')
axes[0].set_ylabel('Frequency')

# Plot for Platelets
sns.histplot(df_cleaned['BMI'].dropna(), ax=axes[1], bins=30, kde=True, color='green')
axes[1].set_title('Distribution of BMI')
axes[1].set_xlabel('BMI')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
